<h3>Applying Linear Regression, Ridge and Lasso on normalized data</h3>

```python
from sklearn.model_selection import train_test_split
y=df1.mpg
x=df1.drop(['mpg'],axis=1)
x.info()
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x)
x_scaled1=pd.DataFrame(x_scaled,columns=x.columns)

y_scaled=preprocessing.scale(y)
y_scaled1=pd.DataFrame(y_scaled)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled1,y_scaled1, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
min_mse=float('inf')
min_alpha=0
i=0.1
while(i<1):
    ridge=Ridge(alpha=i)
    ridge.fit(x_train,y_train)
    y_pred_ridge=ridge.predict(x_test)
    mse=mean_squared_error(y_test,y_pred_ridge)
    if(mse<min_mse):
        min_mse=mse
        min_alpha=i
    i=i+0.1
ridge=Ridge(alpha=min_alpha)
ridge.fit(x_train,y_train)
print(ridge.intercept_)
print(ridge.coef_)
```

```python
from sklearn.linear_model import Lasso
min_mse=float('inf')
min_alpha=0
i=0.1
while(i<1):
    lasso=Lasso(alpha=i)
    lasso.fit(x_train,y_train)
    y_pred_lasso=lasso.predict(x_test)
    mse=mean_squared_error(y_test,y_pred_lasso)
    if(mse<min_mse):
        min_mse=mse
        min_alpha=i
    i=i+0.1
lasso=Lasso(alpha=min_alpha)
lasso.fit(x_train,y_train)
print(lasso.intercept_)
print(lasso.coef_)
```

<h3> Part (d) Implement RidgeCV and LassoCV</h3>

<h4>LassoCV</h4>

```python
from sklearn.linear_model import LassoCV
from sklearn.datasets import make_regression
lcv=LassoCV(cv=5, random_state=0).fit(x_trainn,y_trainn)
print(lcv.score(x_trainn,y_trainn))
```

```python
ypred_lasso=lcv.predict(x_testn)
print(mean_squared_error(y_testn,ypred_lasso))
```

```python
for i,col_name in enumerate(x_trainn.columns):
    print("The coefficient for {} is {}".format(col_name,lcv.coef_[i]))
```

<h4>RidgeCV</h4>

```python
from sklearn.linear_model import RidgeCV
rcv=RidgeCV(alphas=[1e-3,1e-2,1e-1]).fit(x_trainn,y_trainn)
print(rcv.score(x_trainn,y_trainn))
```

```python
ypred_ridge=rcv.predict(x_testn)
print(mean_squared_error(y_testn,ypred_ridge))
```

```python
for i,col_name in enumerate(x_trainn.columns):
    print("The coefficient for {} is {}".format(col_name,rcv.coef_[i]))
```

<h3>Comparing Coefficients</h3>

```python
for i,col_name in enumerate(x_trainn.columns):
    print("The LinearRegression coefficient for {} is {}".format(col_name,lrn.coef_[i]))
    print("The RidgeCV coefficient for {} is {}".format(col_name,rcv.coef_[i]))
    print("The LassoCV coefficient for {} is {}".format(col_name,lcv.coef_[i]))
```
