```python
X,y=data['data'],data['target']
```

```python
X=StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.33)
```

```python
X_train.shape
```

```python
train_acc=[]
test_acc=[]
pred_acc=[]
depth=[i for i in range(1,15)]
for d in depth:
    DT=DecisionTreeClassifier(max_depth=d)
    DT.fit(X_train,y_train)
    y_pred=DT.predict(X_test)
    pred_acc.append(metrics.accuracy_score(y_test,y_pred))
    train_acc.append(DT.score(X_train,y_train))
    test_acc.append(DT.score(X_test,y_test))
```

```python
plt.plot(depth,train_acc)
plt.plot(depth,test_acc,color="red")
plt.plot(depth,pred_acc,linestyle=(0,(1,10)),color="yellow")
```

```python
train_acc=[]
test_acc=[]
pred_acc=[]
max_leaf_nodes=[i for i in range(2,61)]
for d in max_leaf_nodes:
    DT=DecisionTreeClassifier(max_leaf_nodes=d)
    DT.fit(X_train,y_train)
    y_pred=DT.predict(X_test)
    pred_acc.append(metrics.accuracy_score(y_test,y_pred))
    train_acc.append(DT.score(X_train,y_train))
    test_acc.append(DT.score(X_test,y_test))
```

```python
plt.plot(max_leaf_nodes,train_acc)
plt.plot(max_leaf_nodes,test_acc,color="red")
plt.plot(max_leaf_nodes,pred_acc,linestyle=(0,(1,10)),color="yellow")
plt.xlabel("max_leaf_nodes")
plt.ylabel("Accuracy")
```

```python
be=DecisionTreeClassifier(max_depth=5, max_leaf_nodes=7)
be.fit(X_train,y_train)
```

```python
from sklearn.ensemble import BaggingClassifier,AdaBoostClassifier
```

```python
lst=[i for i in range(1,50)]
bag_score=[]
for i in range(1,50):
    bag_model=BaggingClassifier(base_estimator=be,n_estimators=i)
    bag_model.fit(X_train,y_train)
    y_pred=bag_model.predict(X_test)
    bag_score.append(metrics.accuracy_score(y_test,y_pred))
```

```python
plt.plot(lst,bag_score,color="red")
plt.xlabel("n_estimators")
plt.ylabel("Accuracy")
plt.show()
```

```python
ada_score=[]
for i in range(1,50):
    ada_model=AdaBoostClassifier(base_estimator=be,n_estimators=i)
    ada_model.fit(X_train,y_train)
    y_pred=ada_model.predict(X_test)
    ada_score.append(metrics.accuracy_score(y_test,y_pred))
```

```python
plt.plot(lst,ada_score,color="red")
plt.xlabel("n_estimators")
plt.ylabel("Accuracy")
plt.show()
```
